## Desafio PIXFORCE

Na pasta 'data' encontra-se 6 tuplas de arquivos de final ('_trainX.npy','_trainY.npy'). Cada arquivo trainX possui imagens com alvos verdadeiros e falsos e o arquivo trainY contém os respectivos *labels*.

Onde:

    [ 1.,  0.] -> False 

    [ 0.,  1.] -> True 
    
    
### Objetivo:


Carregar, visualizar os dados de forma clara. Criar um classificador para o conjunto total de dados, estabelecer uma métrica para avaliação e apresentar os resultados. 


*Abaixo segue um código básico para simples conferência do dataset, e visualização de duas amostras*

```python
import numpy as np
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
X_files = sorted([join('./data/', f) for f in listdir('./data/') if isfile(join('./data/', f)) and 'trainX' in f])
Y_files = sorted([join('./data/', f) for f in listdir('./data/') if isfile(join('./data/', f)) and 'trainY' in f])
X_files,Y_files
```

```python
for fx,fy in zip(X_files,Y_files):
    sample_x = np.load(fx)
    print('File',fx,'shape:',sample_x.shape)
    sample_y = np.load(fy)
    print('File',fy,'shape:',sample_y.shape)
```

```python
## AMOSTRA NEGATIVA
plt.figure()
plt.imshow(sample_x[1]/255)
plt.show
sample_y[1]
```

```python
## AMOSTRA POSITIVA
plt.figure()
plt.imshow(sample_x[-1]/255)
plt.show
sample_y[-1]
```
